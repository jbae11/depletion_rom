# BU vs Pu for varying enrichments

```python
df = pd.read_csv('all.csv', index_col=0)
df = df.fillna(0)
```

```python
df
```

```python
enr = np.sort(df.init_enr.unique())
bu = np.sort(df.bu.unique())
```

```python
for e in enr:
    y = []
    for b in bu:
        pu239 = list(df[(df['init_enr'] == e) & (df['bu'] == b)]['pu-239'])
        y.append(pu239[0])
    plt.scatter(bu, y)
    plt.xlabel('Burnup')
    plt.ylabel('Concentration [%]')
    plt.title('Enrichment: %s'%str((round(e, 2))))
    plt.savefig('./images/pu239/%s' %str((round(e, 2))).replace('.', '_'))
    plt.show()
    plt.close()
```
