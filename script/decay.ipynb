```python
from pyne import material
from pyne import nucname
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame.from_csv('../curated.csv')
df
```

```python
x = df.iloc[0]
d = {k:v for k,v in x.to_dict().items() if '-' in k}
mat = material.Material(d)
prev_isos = list(df)[5:]
```

```python
d_list = []
for i in range(len(df)):
#for i in range(5):
    x = df.iloc[i]
    d = {k:v for k,v in x.to_dict().items() if '-' in k}
    mat = material.Material(d)
    for _ in range(10):
        # 0 to 50 years
        rand_decay_time = np.random.uniform(0, 50*365)
        # to seconds
        rand_decay_time_secs = rand_decay_time * 3600 * 24
        c = mat.decay(rand_decay_time_secs)
        c = {nucname.serpent(k).lower():v for k,v in c.items()}
        new_dict = {k:v for k, v in c.items() if k in prev_isos}
        new_dict['other'] = sum(new_dict.values())
        new_dict['cooling_time'] = rand_decay_time
        for k in list(df)[:5]:
            new_dict[k] = x.to_dict()[k]
        d_list.append(new_dict)
    if i % 10000 == 0:
        print(i/244896 * 100, '%% done')
```

```python
new_df = pd.DataFrame(d_list)
```

```python
new_df.to_csv('curated_with_decay.csv')
```
